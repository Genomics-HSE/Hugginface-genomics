```python
from math import log, exp

Ne = [1., 0.5, 2.]
t = [0.1, 1.]
eps = 1e-1

def p_nc(ne, t):
    return(exp(-t/ne) )

def ne(pnc, t):
    pnc = max(eps, pnc)
    return(-t/log(pnc))

def prob2size(p, t):
    pnc = 1.
    Ne = []
    for i in range(len(t)):
        Ne.append(ne(1.0-p[i]/pnc,t[i]))
        pnc *= p_nc(Ne[-1],t[i])
    return(Ne)
```
